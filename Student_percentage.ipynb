# Step-1 : Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Taking Raw Data to create a Data Frame

```python
data = {
    "Hours":[2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8],
    "Scores" : [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86]
}
```

```python
# creating a dataframe from raw data
std_data = pd.DataFrame(data)
```

```python
# checking the info to know about data
std_data.info()
```

```python
#checking the data shape
std_data.shape
```

```python
std_data.describe()
```

# Data Visualization

```python
# creating scatter plot to observe the data
plt.figure(figsize=(8, 6))
plt.scatter(std_data["Hours"], std_data['Scores'], color='red', marker='o')
plt.title('Scores vs Hours')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(std_data["Hours"], std_data["Scores"], color='green',width = 0.5)
plt.title('Scores vs Hours')
plt.xlabel('Hours')
plt.ylabel('Scores')
  # Set y-axis ticks based on scores range
plt.grid(axis='y',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# checking the corelation bitween the hours and scores columns
std_data.corr()
```

```python
# Plotting corelation bitween hours and scores columns
plt.figure(figsize=(8, 6))
sns.regplot(x='Hours', y='Scores', data=std_data, color='blue', scatter_kws={'s': 100})
plt.title('Correlation Plot between Hours and Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')

# Calculate and annotate correlation coefficient
corr_coefficient = std_data['Hours'].corr(std_data['Scores'])
plt.text(0.5, 70, f'Correlation Coefficient: {corr_coefficient:.2f}', fontsize=12)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# dividing the  data
X = std_data[["Hours"]]
y = std_data["Scores"]
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
# creating the linear regression model
model = LinearRegression()
```

```python
#fitting the data to model
model.fit(X,y)
```

- predicted score if a student studies for 9.25 hours/day

```python
# predicting
model.predict([[9.25]])
```
